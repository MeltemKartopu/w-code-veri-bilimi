```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
url = r'C:/Users/LENOVO/Desktop/w-code/w-code datasetleri/googleplaystore.csv'
```

```python
df = pd.read_csv(url)
```

```python
print("Veri Seti Sütunları:", df.columns)
print("Eksik Değerler:", df.isnull().sum())
```

```python
 # eksik değerlerin aynı sütundaki medianları (veri setinin ortanca değeri) ile doldurma
df["Rating"].fillna(df["Rating"].median(), inplace=True)
```

```python
# App sütununa göre gruplandırılır ve her bir App değeri için toplanır.
# Gruplandıktan sonra Rating sütunu seçilir.Rating değerlerinin ortalaması alınır 
# Ardından her appe göre ortalama rating hesaplanır. 
df.groupby("App")[["Rating"]].mean().reset_index()
```

```python
df.groupby("Category")[["Rating"]].mean().reset_index()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
numerical_cols = df.select_dtypes(include=['number']).columns

print(categorical_cols)
print(numerical_cols)
```

```python
# Category sütununda olağan dışı değer olan "1.9" kaç tane satırda var 
df[df['Category'] == '1.9']
```

```python
# Category sütununda başka sayısal olmayan değer var mı 
df[df['Category'].str.replace('.', '', 1).str.isdigit()] 
```

```python
df = df[df['Category'] != '1.9'].reset_index(drop=True) 
# Category sütunundan '1.9' içeren satırları kaldırma ve yeniden indexleme
```

# Veri Görselleştirme
 

## Bar Grafiği

```python
sns.color_palette("pastel")
category_rating = df.groupby("Category")[["Rating"]].mean().reset_index()

plt.figure(figsize=(28,6))
plt.bar(category_rating["Category"], category_rating["Rating"])
plt.xlabel("Category")
plt.ylabel("Rating Scores")
plt.title("Category by Rating Scores")
plt.xticks(rotation=90)

```

```python
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
plt.xlabel("Category")
plt.ylabel("Rating Scores")
sns.barplot(x=category_ratings.index, y=category_ratings.values)
plt.xticks(rotation=90)
plt.title('Kategorilere Göre Ortalama Derecelendirmeler')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Type', data=df)
plt.title('Ücretsiz ve Ücretli Uygulama Sayısı')
plt.show()
```

```python
# kategori ve ratinge göre genres (türler) sütun dağılım grafiği 
game_genre_ratings = df[df['Category'] == 'GAME']

plt.figure(figsize=(12, 6))
sns.barplot(x='Genres', y='Rating', data=game_genre_ratings)
plt.xticks(rotation=90)
plt.title('GAME Kategorisi İçindeki Genres\'ların Ortalama Rating\'leri')   
plt.show()
```

## Histogram Grafiği

```python
# Uygulamalara ait derecelerin histogram grafiği
plt.figure(figsize=(8, 6))
sns.histplot(x='Rating', data=df)
plt.title('Uygulama Derecelendirmelerinin Dağılımı')
plt.show()
```

```python
# İndirme sayısı ile derecelendirme arasındaki ilişki Dağılım Grafiği

install_rating = df.groupby("Installs")[["Rating"]].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Installs', y='Rating', data=df)
plt.title('İndirme Sayısı ile Derecelendirme Arasındaki İlişki')
plt.xlabel('İndirme Sayısı')
plt.ylabel('Derecelendirme')
plt.xticks(rotation=90)
plt.show()
```

## Kutu Dağılım (Box Plot)

```python
# Ücretsiz ve ücretli uygulama ortalama derecelendirme box plot grafiği
plt.figure(figsize=(6, 6))
sns.boxplot(x='Type', y='Rating', data=df)
plt.title('Ücretsiz ve Ücretli Uygulamaların Ortalama Derecelendirme Karşılaştırması')
plt.show()
```

## Pasta Dağılım Grafiği (Pie Chart)

```python
# Uygulama kategorilerine göre uygulama sayısı
category_counts = df['Category'].value_counts()

# Pasta grafiği 
plt.figure(figsize=(12, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=35,
        textprops={'fontsize': 9})
plt.title('Uygulama Kategorilerine Göre Dağılım')
plt.axis('equal')  # Daire şeklini sağlamak için
plt.tight_layout()
plt.show()
```

```python
# Art and Design kategorisinde 1000'den fazla yorum alan uygulamaları filtreleme

df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
category_reviews = df[(df['Category'] == 'ART_AND_DESIGN') & (df['Reviews'] > 1000)].reset_index(drop=True)
category_reviews.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', hue='Type', data=df)
plt.title('Kategorilere Göre Ücretli ve Ücretsiz Uygulama Sayıları')
plt.xticks(rotation=90)
plt.xlabel('Kategori')
plt.ylabel('Uygulama Sayısı')
plt.legend(title="Uygulama Türü")  # Legend ekleniyor
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df[df['Type'] == 'Free']['Rating'], label='Ücretsiz', shade=True)
sns.kdeplot(df[df['Type'] == 'Paid']['Rating'], label='Ücretli', shade=True)
plt.title('Ücretli ve Ücretsiz Uygulamaların Rating Dağılımı')
plt.xlabel('Derecelendirme')
plt.ylabel('Yoğunluk')
plt.legend(title="Uygulama Türü")  # Legend ekleniyor
plt.show()
```
