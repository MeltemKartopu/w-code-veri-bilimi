# Veri G√∂rselle≈ütirme
 

## Bar Grafiƒüi

```python
sns.color_palette("pastel")
category_rating = df.groupby("Category")[["Rating"]].mean().reset_index()

plt.figure(figsize=(28,6))
plt.bar(category_rating["Category"], category_rating["Rating"])
plt.xlabel("Category")
plt.ylabel("Rating Scores")
plt.title("Category by Rating Scores")
plt.xticks(rotation=90)

```

```python
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
plt.xlabel("Category")
plt.ylabel("Rating Scores")
sns.barplot(x=category_ratings.index, y=category_ratings.values)
plt.xticks(rotation=90)
plt.title('Kategorilere G√∂re Ortalama Derecelendirmeler')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Type', data=df)
plt.title('√úcretsiz ve √úcretli Uygulama Sayƒ±sƒ±')
plt.show()
```

```python
# kategori ve ratinge g√∂re genres (t√ºrler) s√ºtun daƒüƒ±lƒ±m grafiƒüi 
game_genre_ratings = df[df['Category'] == 'GAME']

plt.figure(figsize=(12, 6))
sns.barplot(x='Genres', y='Rating', data=game_genre_ratings)
plt.xticks(rotation=90)
plt.title('GAME Kategorisi ƒ∞√ßindeki Genres\'larƒ±n Ortalama Rating\'leri')   
plt.show()
```

## Histogram Grafiƒüi

```python
# Uygulamalara ait derecelerin histogram grafiƒüi
plt.figure(figsize=(8, 6))
sns.histplot(x='Rating', data=df)
plt.title('Uygulama Derecelendirmelerinin Daƒüƒ±lƒ±mƒ±')
plt.show()
```

```python
# ƒ∞ndirme sayƒ±sƒ± ile derecelendirme arasƒ±ndaki ili≈üki Daƒüƒ±lƒ±m Grafiƒüi

install_rating = df.groupby("Installs")[["Rating"]].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Installs', y='Rating', data=df)
plt.title('ƒ∞ndirme Sayƒ±sƒ± ile Derecelendirme Arasƒ±ndaki ƒ∞li≈üki')
plt.xlabel('ƒ∞ndirme Sayƒ±sƒ±')
plt.ylabel('Derecelendirme')
plt.xticks(rotation=90)
plt.show()
```

## Kutu Daƒüƒ±lƒ±m (Box Plot)

```python
# √úcretsiz ve √ºcretli uygulama ortalama derecelendirme box plot grafiƒüi
plt.figure(figsize=(6, 6))
sns.boxplot(x='Type', y='Rating', data=df)
plt.title('√úcretsiz ve √úcretli Uygulamalarƒ±n Ortalama Derecelendirme Kar≈üƒ±la≈ütƒ±rmasƒ±')
plt.show()
```

## Pasta Daƒüƒ±lƒ±m Grafiƒüi (Pie Chart)

```python
# Uygulama kategorilerine g√∂re uygulama sayƒ±sƒ±
category_counts = df['Category'].value_counts()

# Pasta grafiƒüi 
plt.figure(figsize=(12, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=35,
        textprops={'fontsize': 9})
plt.title('Uygulama Kategorilerine G√∂re Daƒüƒ±lƒ±m')
plt.axis('equal')  # Daire ≈üeklini saƒülamak i√ßin
plt.tight_layout()
plt.show()
```

```python
# Art and Design kategorisinde 1000'den fazla yorum alan uygulamalarƒ± filtreleme

df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
category_reviews = df[(df['Category'] == 'ART_AND_DESIGN') & (df['Reviews'] > 1000)].reset_index(drop=True)
category_reviews.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', hue='Type', data=df)
plt.title('Kategorilere G√∂re √úcretli ve √úcretsiz Uygulama Sayƒ±larƒ±')
plt.xticks(rotation=90)
plt.xlabel('Kategori')
plt.ylabel('Uygulama Sayƒ±sƒ±')
plt.legend(title="Uygulama T√ºr√º")  # Legend ekleniyor
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df[df['Type'] == 'Free']['Rating'], label='√úcretsiz', shade=True)
sns.kdeplot(df[df['Type'] == 'Paid']['Rating'], label='√úcretli', shade=True)
plt.title('√úcretli ve √úcretsiz Uygulamalarƒ±n Rating Daƒüƒ±lƒ±mƒ±')
plt.xlabel('Derecelendirme')
plt.ylabel('Yoƒüunluk')
plt.legend(title="Uygulama T√ºr√º")  # Legend ekleniyor
plt.show()
```
